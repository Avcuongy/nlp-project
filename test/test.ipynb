```python
from joblib import load
import pandas as pd
import numpy as np
```

```python
# Load vectorizer (saved with joblib, not pickle)
vectorizer = load("../models/vectorizer.pkl")

print("Vectorizer loaded successfully!")
print(f"Vectorizer type: {type(vectorizer)}")
print(f"Vocabulary size: {len(vectorizer.vocabulary_)}")
```

```python
# Load train.csv
df_train = pd.read_csv("../data/processed/train.csv")

print(f"Train data shape: {df_train.shape}")
print(f"\nFirst few rows:")
print(df_train.head())

# Transform the text data using vectorizer
X_train = vectorizer.transform(df_train["comment"])

print(f"\nTransformed data shape: {X_train.shape}")
print(f"Data type: {type(X_train)}")
print(f"Sparse matrix density: {X_train.nnz / (X_train.shape[0] * X_train.shape[1]):.4f}")
```
